```python
df2 = df1[["RCP_PARTS_DTLS", "RCP_NM"]]

# split the recipe parts by the newline character
df2['recipe_parts_list'] = df2['RCP_PARTS_DTLS'].str.split('\n')

# split the recipe parts by comma
df2 = df2.explode('recipe_parts_list')
df2[['ingredients', 'measurements']] = df2['recipe_parts_list'].str.split(',', n=1, expand=True)

# split the ingredients into name, quantity, and unit
df2[['재료', '함량']] = df2['ingredients'].str.rsplit(' ', n=1, expand=True)
df2['단위'] = df2['함량'].str.extract('(\D+)')
df2['함량'] = df2['함량'].str.extract('(\d+)')
df2.drop(columns=['ingredients', 'recipe_parts_list'], inplace=True)

# split the measurements into amount and cup measurement
df2[['amount', '목측량']] = df2['measurements'].str.split('(', n=1, expand=True)
df2['목측량'] = df2['목측량'].str.extract('(\d.*\))')
df2.drop(columns=['measurements'], inplace=True)

# add the recipe name and stage columns
df2['단계'] = df2["RCP_PARTS_DTLS"].str.split('\n', n=1).str[0]
df2['레시피명'] = df2["RCP_NM"]

# rearrange the columns
df3 = df2[['레시피명', '단계', '재료', '함량', '단위', '목측량']]

df3.head(10)
```

```python
# define a regular expression pattern to match ingredient lines
pattern = r'^([^0-9]+)\s*([0-9]+)g?\s*\(([^\)]+)\),?\s*(.*)'

# create an empty list to hold the parsed ingredient data
ingredients = []

# iterate over each line of the original data
for line in df2["RCP_PARTS_DTLS"].astype(str):

    # use the regular expression pattern to match the line
    match = re.match(pattern, line)
    
    # if the line matches the pattern, extract the data
    if match:
        recipe_name = df2.loc[df2["RCP_PARTS_DTLS"]==line].index
        step_name = match.group(1).strip()
        print(step_name)
        ingredient_name = match.group(3).strip()
        print(ingredient_name)
        amount = match.group(2).strip()
        print(amount)
        unit = 'g'
        estimation = match.group(4).strip()
        print(estimation)
        
        # append the parsed data to the list of ingredients
        ingredients.append([recipe_name, step_name, ingredient_name, amount, unit, estimation])

# create a pandas dataframe from the list of ingredients
df3 = pd.DataFrame(ingredients, columns=['레시피명', '단계', '재료', '함량', '단위', '목측량'])

# fill in missing step names by copying the previous step name
df3['단계'] = df3['단계'].fillna(method='ffill')

# fill in missing estimation values with empty strings
df3['목측량'] = df3['목측량'].fillna('')

# print the resulting dataframe
df3.head(15)
```
