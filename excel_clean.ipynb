```python
import pandas as pd
import re
```

```python
df1 = pd.read_excel("OJT_raw_data.xlsx")
df1.drop(['Unnamed: 0'], axis = 1, inplace = True)
df1
```

```python
df2 = df1[["RCP_PARTS_DTLS", "RCP_NM"]]
for index in range(len(df2)):
    print('**' + df2.iloc[index]['RCP_NM'] + '**')
    print(df2.iloc[index]['RCP_PARTS_DTLS'])
    print('================================')
```

```python

refined_data = []

recipe_name = df2['RCP_NM']

for line in df2["RCP_PARTS_DTLS"].astype(str):
    line2 = line.replace('\n', ' ,')
    ingredients = re.findall(r'([\w\(\)\s]+)\s([약간|\d.]+)?([g|mg|ml|Ts|큰술|작은술|약간]+)?[(]?([\d\w\/×\s?]+)?[)]?(?:\n[\w\s]+)?(?:\n[\w\s]+)?,?\s*', line2)
    item1 = "주재료"
    for ingredient in ingredients:
        item2 = ingredient[0]
        item3 = ingredient[1]
        item4 = ingredient[2]
        item5 = ingredient[3]
        if ingredient[1] == '' and ingredient[2] == '' and ingredient[3] == '':
            if '약간' in ingredient[0]:
                ingr_stripped = ingredient[0].strip()
                if ' ' in ingr_stripped:
                    ingr_split = ingr_stripped.split(' ')
                    item2 = ingr_split[0]
                    item3 = ingr_split[-1]
                else:
                    item2 = ingr_stripped.replace('약간', '')
                    item3 = '약간'
            else:
                item1 = ingredient[0]
                continue
        refined_data.append({'레시피명': df2.loc[df2[df2["RCP_PARTS_DTLS"]==line].index, 'RCP_NM'].to_string(index=False),
                             '단계': item1, 
                             '재료': item2, 
                             '함량': item3, 
                             '단위': item4, 
                             '목측량': item5})

refined_df = pd.DataFrame(refined_data)

refined_df.head(50)
```

```python
refined_df.loc[291:297, "단계"] = "구운채소"
refined_df.loc[325:328, "단계"] = "닭가슴살롤"
refined_df.loc[498:502, "단계"] = "소스"
refined_df.loc[640:642, "단계"] = "육수"
refined_df.loc[700:702, "단계"] = "고구마무스"
refined_df.loc[1029:1033, "단계"] = "육수"
```

```python
final_df1 = refined_df.loc[:1199]
final_df1.to_csv("refined_recipes1.csv")
```
